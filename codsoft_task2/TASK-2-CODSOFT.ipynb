```python
def stylish_line(text=""):
    print("‚îÄ" * 45)
    if text:
        print(text)
        print("‚îÄ" * 45)

def calculate(a, b, op):
    operations = {
        "+": lambda x, y: x + y,
        "-": lambda x, y: x - y,
        "*": lambda x, y: x * y,
        "/": lambda x, y: "‚ùå Cannot divide by zero!" if y == 0 else x / y
    }
    return operations.get(op, None)(a, b) if op in operations else None

while True:
    stylish_line(" BASIC ARITHMETIC CALCULATOR üî¢ ")

    try:
        num1 = float(input("Enter first number  ‚û§ "))
        num2 = float(input("Enter second number ‚û§ "))
    except ValueError:
        stylish_line("‚ùå Invalid input! Please enter numbers only.")
        continue

    stylish_line("Choose Operation:  +   -   *   /")
    op = input("Enter operation       ‚û§ ").strip()

    if op not in ["+", "-", "*", "/"]:
        stylish_line("‚ùå Invalid input! Please enter above operations only.")
        continue

    result = calculate(num1, num2, op)

    stylish_line(" RESULT ")
    print(f" {num1}  {op}  {num2}  =  {result}")
    stylish_line(" Thank you for using this calculator! ")

    exit_now = False
    while True:
        again = input("Do you want to perform another calculation? (yes/no) ‚û§ ").strip().lower()
        if again in ["yes", "y"]:
            break
        elif again in ["no", "n"]:
            stylish_line(" Exiting calculator. Thank you for using it! ")
            exit_now = True
            break
        else:
            stylish_line("‚ùå Invalid input! Please enter yes/no only.")
    if exit_now:
        break
```
