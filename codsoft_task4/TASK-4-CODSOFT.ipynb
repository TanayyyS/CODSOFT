```python
import random
import time
import os
import sys

def clear():
    try:
        os.system('cls' if os.name == 'nt' else 'clear')
    except Exception:
        pass

choices = ["rock", "paper", "scissors"]

user_score = 0
computer_score = 0
round_number = 1
game_running = True

while game_running:
    clear()
    print(f"\n===== ROUND {round_number} =====")
    print("Rock ‚Ä¢ Paper ‚Ä¢ Scissors")
    print("---------------------------")

    user = input("Your choice: ").strip().lower()
    if user not in choices:
        print("\nInvalid choice! Please choose rock, paper, or scissors.")
        time.sleep(1.5)
        continue

    print("\nComputer is thinking", end="")
    for _ in range(3):
        print(".", end="", flush=True)
        time.sleep(0.4)
    print()

    computer = random.choice(choices)

    print(f"\nYou chose     : {user.upper()}")
    print(f"Computer chose: {computer.upper()}")
    print("---------------------------")

    if user == computer:
        print("Result: It's a TIE!")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("Result: YOU WIN üéâ")
        user_score += 1
    else:
        print("Result: YOU LOSE ‚ùå")
        computer_score += 1

    print(f"\nScoreboard ‚Üí You: {user_score} | Computer: {computer_score}")
    print("---------------------------")

    while True:
        again = input("\nPlay again? (yes/no): ").strip().lower()
        if again == "yes":
            round_number += 1
            time.sleep(0.6)
            break
        elif again == "no":
            print("\nExiting Rock-Paper-Scissors game! Thank you for playing.")
            print(f"Final Score ‚Üí You: {user_score} | Computer: {computer_score}")
            game_running = False
            break
        else:
            print("Invalid input! Please enter yes/no only.")
```
